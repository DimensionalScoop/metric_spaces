```python
import numpy as np
import matplotlib.pyplot as plt
from empiricaldist import Pmf
```

```python
q = np.linspace(0,0.5)
p = np.array([1]*len(q))

original = Pmf(p, index=q)
original.normalize()


sqrt = Pmf(p, index=np.sqrt(q))
sqrt.normalize()

original.plot()
sqrt.plot()
```

```python
a = np.random.rand(100000)/2
b = np.sqrt(a)

plt.hist(a, bins=100, density=True)
plt.hist(b, bins=100, histtype="step", density=True)
plt.show()
```

```python
a.mean()**2, a.std()**2
```

```python
b.mean()**2, b.std()**2
```

```python
a.mean()**2/b.mean()**2 * b.std()**2 / a.std()**2
```
